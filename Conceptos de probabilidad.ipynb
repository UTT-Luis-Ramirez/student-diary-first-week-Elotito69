#CONCEPTOS DE PROBABILIDAD
**Probabilidad**: Es un área de las matemáticas que estudia la probabilidad de que ocurran ciertos eventos. Es importante aclarar que no existe ningún evento con una probabilidad del 100% de que suceda o del 0% de que no suceda. Además, nos ayuda a entender de manera matemática qué tan probable es que ocurra un evento. El objetivo de la probabilidad es medir y analizar la incertidumbre de los eventos o situaciones que pueden ocurrir en el futuro.

**Tipos de probabilidad**


*   Probabilidad a priori: Se refiere a la probabilidad de un evento      antes de que ocurra otro evento.
*   Probabilidad a posteriori:Se refiere a la probabilidad de que ocurra un evento después de que haya sucedido otro evento."


*   Probabilidad mutuamente exclusiva: Probabilidad de que dos eventos ocurran al mismo tiempo.
Esto por decir algunos ejemplos de probabilidad ahora continuaremos con el concepto de estadistica.

**Estadística**: La estadística es una rama que estudia el comportamiento de los datos, además de que transforma los datos en información para tomar decisiones a futuro.

**Tipos de estadística**


*   Estadística descriptiva
*   Estadística inferencial

Aunque los conceptos de probabilidad y estadística están relacionados, para que cada uno de ellos quede más claro, Una de las diferencias más notorias es su uso, ya que la probabilidad se utiliza para predecir resultados a futuro y la estadística se utiliza para analizar datos recolectados de encuestas o investigaciones.

 **Teoría de conjuntos**: Es una rama de las matemáticas que estudia los conjuntos, es decir, colecciones de objetos o elementos. Estos objetos pueden ser cualquier tipo de entidad, como números, letras, personas, etc. La teoría de conjuntos nos ayuda a organizar grupos bien definidos y nos facilita la resolución de problemas al dividirlos para llegar a un resultado.

*Dato*: Información sobre un evento que ya ha ocurrido.

*Evento*: Un resultado de un suceso que aún no ha ocurrido.

*Axiomas*: Afirmaciones que se consideran correctas hasta que se demuestre lo contrario.

Falacias: Afirmaciones incorrectas hasta que se demuestre lo contrario.

#Estadística Descriptiva
Trata de mostrar la información de manera fácil y sencilla, y busca la manera de interpretar datos. La secuencia de pasos en que se realiza la estadística descriptiva es la siguiente:

*   Recolección y organización de los datos.
*   Organizar los datos en tablas o gráficas.
*   Cálculo de medidas descriptivas
*   Identificación de patrones o operaciones.

**Cálculo de medidas descriptivas:** Media, mediana y moda.

**Medidas de Dispersión:** Rango, varianza y desviación estándar.

**Medidas Simétricas:** Asimetría y curtosis.

**Aplicaciones de la estadística descriptiva:** Lo utilizamos para finanzas personales, resultados deportivos, pronósticos del clima y en la medicina.

#Estadística inferencial

Se encarga de extraer conclusiones a través de una muestra. Busca una proyección a futuro. Necesitamos la población para crear una hipótesis, que es una suposición de algo. Y hay que ser honestos con nuestras propias afirmaciones.
Generar un porcentaje de validación de la hipótesis.

**Modelado estadístico:** Inferencia futurista de cómo ocurriría algo basándonos en el intervalo de confianza.

**Aplicaciones de la estadística inferencial:** Para el control de calidad.
Estudios médicos para describir el tipo de patrones y ver cómo afectaba la enfermedad. Investigación científica, marketing e inversiones. Ciencias sociales.
